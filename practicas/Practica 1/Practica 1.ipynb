```python
def obtener_libro(id = 40373):
    raw_book = gutenbergpy.textget.get_text_by_id(id) # with headers
    clean_book = gutenbergpy.textget.strip_headers(raw_book) # without headers
    return clean_book, raw_book
```

```python
import re

# Si libro_filtrado es una lista de cadenas
if isinstance(libro_filtrado, list):
    texto_limpio = [re.sub(r'[^\w\s]', '', texto.lower()) for texto in libro_filtrado]
else:
    # Si es una sola cadena
    texto_limpio = re.sub(r'[^\w\s]', '', libro_filtrado.lower())

# Muestra el resultado
print(texto_limpio)
```

```python
# Eliminar stopwords
stop_words = set(stopwords.words('english'))
list(stop_words)[:20]
```

```python
from nltk.tokenize import word_tokenize

# Si texto_limpio es una lista
tokens = [word_tokenize(text) for text in texto_limpio]

print(tokens)
```

```python
from nltk.tokenize import word_tokenize

# Concatenar lista en una sola cadena
texto_concatenado = ' '.join(texto_limpio)

# Tokenizar el texto concatenado
tokens = word_tokenize(texto_concatenado)

print(tokens)
```

```python
# Filtrar solo palabras alfabéticas (sin signos de puntuación ni números)
palabras = [palabra for palabra in tokens if palabra.isalpha()]
```

#### STEMM

```python
# Importar librerías
import nltk
from nltk.stem import PorterStemmer
from collections import Counter
import matplotlib.pyplot as plt
```

```python
plt
# Crear instancia del PorterStemmer
porter = PorterStemme()

# Aplicar stemming a cada palabra
stems = [porter.stem(palabra) for palabra in palabras]
```

```python
frecuencias = Counter(stems)
```

```python
# Mostrar las 10 raíces más comunes
print("Raíces más comunes: \n")
for raiz, frecuencia in frecuencias.most_common(25):
    print(f"{raiz}: {frecuencia}")
```

```python
# Obtener las 10 raíces más comunes
top_raices = frecuencias.most_common(25)
raices, conteos = zip(*top_raices)

# Crear gráfico de barras
plt.figure(figsize = (10,5))
plt.bar(raices, conteos, color = 'lightblue')
plt.title("Top 25 Raíces Más Comunes")
plt.xlabel("Raíces")
plt.ylabel("Frecuencia")
plt.show()
```

#### n-Gramas

```python
from nltk import ngrams
```

```python
n = 2  # Definir el tamaño del n-grama (bigrama)
bigramas = ngrams(palabras, n)  # Pasar directamente la lista

# Imprimir los bigramas
for gramas in bigramas:
    print(gramas)
```

```python
import networkx as nx
n=5
lista_bigramas = ngrams(palabras, n)

G = nx.DiGraph()

for gram in lista_bigramas:
    G.add_edge(gram[0], gram[1])

plt.figure(figsize=(10, 6))
nx.draw(G, with_labels=True, node_size=3000, font_size=14, font_color='orange')
plt.title("Grafo de Bigramas")
plt.show()
```
